```python
import codecs
import json
def get_ipynb_json(filename ):
    """

    Parameters
    ----------
    filename : TYPE
        Takes IPYNB file name 

    Returns
    -------
    y : TYPE
        Json object fo IPYNB file

    """
    
    f = codecs.open(filename, 'r')
    source = f.read()
    
    y = json.loads(source)
    return y

file1, file2 = 'numpan.ipynb', 'numpan2.ipynb'

file1_json =  get_ipynb_json(file1)
file2_json =  get_ipynb_json(file2)
  
print("Comparing Files.........")  

# basic check : for number of cells

print("Checking number of cells.")
x= len(file1_json['cells'] )
y = len(file2_json['cells'] )

print(f"{file1} has '{x}' cells and {file2} has '{y}' cells")


#remove empty cells and check
print("Checking number of cells after removing empty cells")
m = len([i for i in file1_json['cells'] if i['source'] ])
n = len([i for i in file2_json['cells'] if i['source'] ])

print(f"{file1} has '{m}' cells and {file2} has '{n}' cells")

#method to compare outputs of files


# def compare_output():
    
cells_list1 = [i for i in file1_json['cells'] if i['source'] ] # list of dictionaries
cells_list2 = [i for i in file2_json['cells'] if i['source'] ]

for index in range(len(cells_list1)):
    # break
    cellout_file1 = cells_list1[index]['outputs'][0]['data']['text/plain']
    cellout_file2 = cells_list2[index]['outputs'][0]['data']['text/plain']
    
    cellout_file1 == cellout_file2
    
    
    numcommon_lines = len(set(cellout_file1).intersection(set(cellout_file2)))
    
    
    print(f"cell {index} Output Similarity wrt file1 in % : {(numcommon_lines/ len(cellout_file1))*100:.2f}%")
    
    print(f"cell {index} Output Similarity wrt file2 in % : {(numcommon_lines/ len(cellout_file2))*100:.2f}%")
    
    
    
 # Numpy comp :  "array([0, 1, 2, 3, 4, 5])" == "array([0, 1, 2, 3, 4, 5, 6])"   
    
    
      
    
    
    
    
    
    
    
    
    
    
    
    
    
    
 

    
    
    
    
    
    
    
   
    
    
    
    
    
    
```
